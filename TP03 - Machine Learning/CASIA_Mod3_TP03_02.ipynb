1. Charger les données

```sql
%python

train_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("/Volumes/ngow_lakehouse/ml_sandbox/data/train.csv")

test_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("/Volumes/ngow_lakehouse/ml_sandbox/data/test.csv")
```

2. Aperçu général des données

```sql
%python

train_df.show(5)
train_df.printSchema()
train_df.describe().show()
```

3. Vérifier les valeurs manquantes

```sql
%python

from pyspark.sql.functions import col, sum

missing_train = train_df.select(
    [(sum(col(c).isNull().cast("int"))).alias(c) for c in train_df.columns]
)

missing_train.show()
```

```sql
%python
# Convertir en Pandas pour visualiser (recommandé)
train_pd = train_df.toPandas()


# 4. Visualiser les distributions (histogrammes)
import matplotlib.pyplot as plt
import seaborn as sns

train_pd.hist(figsize=(12, 10))
plt.tight_layout()
plt.show()

# 5. Matrice de corrélation
plt.figure(figsize=(12, 8))
sns.heatmap(train_pd.select_dtypes(include=['number']).corr(), annot=False, cmap="viridis")
plt.title("Matrice de corrélations")
plt.show()

# 6. Distribution de la variable cible (si connue)
sns.histplot(train_pd["Transported"], kde=True)
plt.title("Distribution de la variable cible")
plt.show()

# 7. Relations entre variables (pairplot)
sns.pairplot(train_pd.select_dtypes(include=['float64', 'int64']).sample(500))
plt.show()
```
