```sql
/*
Produits actifs :
Récupérez tous les produits encore en
vente (SellEndDate IS NULL) et triés par
leur prix décroissant.
(Table : Product)
*/
SELECT
p.ProductID AS ProduitID,
p.Name AS Nom_Produit,
p.ListPrice AS Prix,
p.sellstartdate AS Date_de_debut,
p.sellenddate AS Date_de_fin
FROM
 Product p
WHERE
 SellEndDate IS NULL
ORDER BY
 ListPrice DESC
```
