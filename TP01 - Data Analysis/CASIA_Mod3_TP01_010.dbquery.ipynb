```sql
/*
Analyse des remises :
Calculez le total des remises
(UnitPriceDiscount) appliquées par
produit et identifiez celui avec la
remise totale la plus élevée.
(Table : SalesOrderDetail)
*/
SELECT 
  sod.ProductID,
  SUM(sod.UnitPrice * sod.OrderQty * sod.UnitPriceDiscount) AS Total_Discount
FROM saleslt.SalesOrderDetail sod
GROUP BY sod.ProductID
ORDER BY SUM(sod.UnitPrice * sod.OrderQty * sod.UnitPriceDiscount) DESC
;
```
