```sql
/*
Commandes par région :
Trouvez le nombre de commandes
pour chaque région (StateProvince),
triées par région.
(Tables : Address, SalesOrderHeader)
*/
SELECT ad.StateProvince as Region, 
COUNT(soh.SalesOrderID) as NbCommandes 
FROM 
saleslt.address ad,
saleslt.salesorderheader soh
WHERE 
ad.AddressID = soh.ShipToAddressID
GROUP BY ad.StateProvince
ORDER BY ad.StateProvince
```
