```sql
/*
Produits par catégorie :
Affichez le nombre total de produits
dans chaque catégorie de produit.
(Tables : Product, ProductCategory)
*/
SELECT
pc.ProductCategoryID,
pc.Name,
COUNT(pd.productid) as NbProduits
FROM
saleslt.product pd,
saleslt.productcategory pc
WHERE
pd.ProductCategoryID = pc.ProductCategoryID
GROUP BY
pc.ProductCategoryID,
pc.Name ORDER BY
COUNT(pd.productid) DESC;
```
