```sql
/*
1. Liste des produits :Récupérez une liste de tous lesproduits avec leurs noms, prix(ListPrice) et catégories.
(Tables : Product, ProductCategory)
*/
SELECT 
pd.name as ProductName,
pc.name as ProductCategoryName,
pd.ListPrice as ListPrice
FROM
 saleslt.product pd, 
 saleslt.productcategory pc 
WHERE pd.ProductCategoryID = pc.ProductCategoryID
```
