```sql
/*
Top 3 clients générant le plus de
revenus :
Identifiez les 3 clients ayant généré le
plus de revenus totaux à travers leurs
commandes.
(Tables : SalesOrderHeader, Customer)
*/
SELECT 
  c.CustomerID,
  SUM(soh.TotalDue) AS RevenuTotal
FROM 
SalesOrderHeader soh,
Customer c
WHERE soh.CustomerID = c.CustomerID
GROUP BY c.CustomerID
ORDER BY SUM(soh.TotalDue) DESC
LIMIT 3;
```
