```sql
/*
Revenus par catégorie de produit :
Calculez les revenus totaux générés
pour chaque catégorie de produit.
(Tables : SalesOrderHeader,
SalesOrderDetail, Product,
ProductCategory)
*/
SELECT 
  pc.Name AS Categorie,
  SUM(soh.TotalDue) AS RevenuTotal
FROM 
saleslt.SalesOrderHeader soh,
saleslt.SalesOrderDetail sod,
saleslt.Product pd,
saleslt.ProductCategory pc
WHERE soh.SalesOrderID = sod.SalesOrderID
AND sod.ProductID = pd.ProductID
AND pd.ProductCategoryID = pc.ProductCategoryID
GROUP BY pc.Name
ORDER BY SUM(soh.TotalDue) DESC;
```
